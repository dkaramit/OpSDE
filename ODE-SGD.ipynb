# Solve ODE using Stochastic Gradient descent

# Example

Let's say we want to solve 
$$
\dfrac{dy}{dt}=-sin(t) \;, \; \text{ with} \;\; y(0)=1 \;.
$$

In order to solve it, we will use a model function:
$$
y=w_6 \ \sum_{i=0}^{2}sin(t +w_i) + w_7 \ \sum_{i=3}^{5}cos(t +w_i) +w_8.
$$

In this example we will use Stochastic Gradient descent to find the values of $\{\bf{w}\}$ that minimize the loss function.

````python
import os

os.chdir('Optimizers/Stochastic-Gradient-Descent/python/')
import SGD as SGD 
os.chdir('../../../')

import numpy as np

import matplotlib
matplotlib.use('nbAgg')
import matplotlib.pyplot as plt

import time
````

Define the model function. Overload the ```modelBase``` class found in ```SGD```.

````python
class modelFunc(SGD.modelBase):
    def __init__(self,dimensions,w0,h=1e-3):
        SGD.modelBase.__init__(self,dimensions,w0)
        
        self.ds_ldx=0 #derivative of l^th signal wrt the input
        self.h=h
        
    def __call__(self):
#         self.signal[0]=np.sin(self.input*self.w[1])*self.w[0]+np.cos(self.input*self.w[3])*self.w[2]
        self.signal[0]=self.w[8]+self.w[6]*sum([np.sin(self.input+self.w[i]) for i in range(3)])+self.w[7]*sum([np.cos(self.input+self.w[i]) for i in range(3,6)])
    
    def derivative_x(self,l):
        _x=self.input
        heff=np.abs(_x)*self.h+self.h
        
        _x-=heff
        self.setInput(_x)
        self()
        f0=self.signal[l]

        _x+=2*heff
        self.setInput(_x)
        self()
        f1=self.signal[l]

        self.ds_ldx=(f1-f0)/(2*heff)

        _x-=heff
        self.setInput(_x)
````

Define the region in which you want to solve the ODE, and the initial condition. Call this class ```Boundary``` because we are going to generalize it to boundary value problems and PDEs.

````python
class Boundary:
    def __init__(self,model,points,y0):
        '''
        model: the model function
        points: the points at which we want the solution
        y0: the value of y at points[0]
        '''
        self.model=model
        self.points=points[:]
        self.y0=y0

        self.NPoints=len(points)
        
        self.boundary_diff=[0 for _ in range(model.dimensions[1])]
        
        self.model.setInput(self.points[0])#start at points[0]
        
    def boundaryDifference(self):
        '''
        get the difference between self.y0 and the output of the model for model.input=points[0]
        '''
        tmp_x=self.model.input
        
        self.model.setInput(self.points[0])
        
        self.model()
        for i,_y in enumerate(self.y0):
            self.boundary_diff[i]=self.model.signal[i] -_y 
            
        self.model.setInput(tmp_x)
        
````

Define the ODE, which basically defines the loss that will be used to train the model. Again, we are goind to generalize this class to be used to PDEs and boundary value problems.

````python
class DifferentialEquation:
    def __init__(self,Model,Boundary,h=1e-3):
        
        #this is the RHS of the ODE. Maybe we should pass it as input, 
        #but since a lot of things depent on the problem, maybe we need a base class and allow overloading
        self.RHS=[lambda x:np.sin(x)]
        
        self.model=Model
        self.boundary=Boundary
        self.h=h
        
        self.data_size=self.boundary.NPoints
        
        self.DE=[0 for _ in range(self.model.dimensions[1])]#this should be LHS-RHS, which we want to be 0 (see definition of __call__)
        
        self.dQdw=0        
        
    def __call__(self):
        for l in range(self.model.dimensions[1]):
            self.model.derivative_x(l)
            self.DE[l]= self.model.ds_ldx + self.RHS[l](self.model.input)
        
        
        
        
    def randomDataPoint(self):
        index=np.random.randint(self.data_size)
        self.x=self.boundary.points[index]    
        self.model.setInput(self.x)
    
    def averageLoss(self):
        '''get the average total loss'''
        
        avrgLoss=0
        
        #loss for the initial condition
        self.boundary.boundaryDifference()
        for l in range(self.model.dimensions[1]):
            avrgLoss+=(self.boundary.boundary_diff[l]**2)
        
        #loss for the derivative at the random point 
        self()
        for l in range(self.model.dimensions[1]):
            avrgLoss+=(self.DE[l]**2)
        
        return avrgLoss/(2.*self.model.dimensions[1])
        

    def grad(self,i):
        '''Get the gradient of the averge loss wrt w[i] at current point + initial conditon'''
        '''we should find a way to do it analytically (at least the dQds part)'''
        
        heff=self.h*np.abs(self.model.w[i])+self.h

        self.model.w[i]-=heff
        Q0=self.averageLoss()

        self.model.w[i]+=2*heff
        Q1=self.averageLoss()

        self.dQdw=(Q1-Q0)/(2.*heff)

        self.model.w[i]-=heff
        
````

````python
model=modelFunc([1,1],np.random.rand(9)*2-1)
boundary=Boundary(model,np.linspace(0,np.pi*2,1500),[1])
ODE=DifferentialEquation(model,boundary)
````

````python
# sgd=SGD.VanillaSGD(ODE,alpha=1e-2)
# sgd=SGD.RMSpropSGD(ODE,gamma=1-1e-2,epsilon=1e-5,alpha=1e-1)
# sgd=SGD.AdaDeltaSGD(ODE,gamma=0.99,epsilon=1e-5,alpha=1)
# sgd=SGD.AdamSGD(ODE,beta_m=0.9,beta_v=0.999,epsilon=1e-8,alpha=1e-1)
# sgd=SGD.AdaMaxSGD(ODE,beta_m=0.9,beta_v=0.999,epsilon=1e-8,alpha=1e-1)
sgd=SGD.NAdamSGD(ODE,beta_m=0.9,beta_v=0.999,epsilon=1e-8,alpha=1e-1)
````

````python
time0=time.time()

sgd.run(abs_tol=1e-3, rel_tol=1e-3, step_break=500,max_step=5000)

time.time()-time0,len(sgd.steps)
````
